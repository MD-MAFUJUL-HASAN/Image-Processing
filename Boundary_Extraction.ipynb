```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read the image
img = cv2.imread('lena.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
# Apply Canny edge detection
edges = cv2.Canny(img, 100, 200)
```

```python
# Create a black image to draw the boundaries on
boundary_img = np.zeros_like(img)
```

```python
# Find contours of the edges
contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
```

```python
# Draw the boundaries on the black image
cv2.drawContours(boundary_img, contours, -1, (255, 255, 255), thickness=1)
```

```python
# Display the original image and the extracted boundaries
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(img, cmap='gray')
axes[0].set_title('Original Image')
axes[0].axis('off')
axes[1].imshow(boundary_img, cmap='gray')
axes[1].set_title('Boundary Extraction')
axes[1].axis('off')
plt.show()
```
