```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read the image
img = cv2.imread('lena.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
# Calculate the negative of the image
neg_img = cv2.bitwise_not(img)
```

```python
# Display the original image and the negative image
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(img, cmap='gray')
axes[0].set_title('Original Image')
axes[0].axis('off')
axes[1].imshow(neg_img, cmap='gray')
axes[1].set_title('Negative Image')
axes[1].axis('off')
plt.show()
```
